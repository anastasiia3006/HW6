```python
#5. Покажіть детальну інформацію про найменш та найбільш прибуткові фільми у наборі даних.

df_max_profitability_isin = df.loc[df['Profitability'].isin([df['Profitability'].max()])]
print("\nНайбільш прибутковий фільм :")
print(df_max_profitability_isin)

# Найменш прибутковий фільм
df_min_profitability = df.loc[df['Profitability'].isin([df['Profitability'].min()])]
print("\nНайменш прибутковий фільм:")
print(df_min_profitability)
```

```python
#6. Значення "Жанр" часом здається непослідовним; спробуйте знайти ці невідповідності та виправити їх

df['Genre'].unique()
```

```python
df['Genre'] = df['Genre'].replace({'Romance': 'Romance', 'romance': 'Romance', 'Romence': 'Romance', 'comedy': 'Comedy', 'Comdy': 'Comedy'})
df['Genre'].unique()
```

```python
#	7. Збережіть (у новий файл CSV) 10 найкращих комедій за кількістю глядачів;  покажіть лише назву фільму, рік та студію.
best_comedy = df['Genre'] == 'Comedy'
best_ten_comedy = df[best_comedy].sort_values(by='Audience score %', ascending=False).head(10)
print(best_ten_comedy[['Film', 'Year', 'Lead Studio']])
```

```python
#	Зберігається результат у новому CSV файлі.
with open('best_comedy.csv', 'w') as f:
    best_ten_comedy[['Film', 'Year', 'Lead Studio']].to_csv(f, index=False)
```
